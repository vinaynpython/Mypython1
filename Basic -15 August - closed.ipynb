## Handy helpful function for # list ## 
- del < To delete the values present in string or list

```python
m = ["First", "Second", 3] ;  # in list the position start as 0,1,2,3,4 -
del m[0] ;                           #  here at 0 its value is "First" and so on
print(m)                            # here we deleteout value present on 0 position
```

```python
m.pop() ; print (m) # Pop is used to delete the specific value present in list 
    # If you do not define value in bracket it pops out last value present in list
```

```python
m.extend("1"); print(m) # Extend only takes one value in list
```

```python
m.append("One"); print(m) # Append also adds the value and even takes one value
```

```python
m.count('1') # Count function use to count the number characters which has repeated in list
```

```python
m.clear();  print(m) ; # Flush out all the values assigned to a variable
```
